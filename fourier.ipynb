```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
a_0 = 1/3
def a(n):
    return 2/(n*np.pi)**2*(-1)**n
def b(n):
    return 0
def y(x : np.ndarray, n : int):
    # This is the sum of the first n terms of the fourier series.
    if n == 0:
        return np.ones(x.shape)*a_0
    else:
        return y(x,n-1) + a(n)*np.cos(n*np.pi*x) + b(n)*np.sin(n*np.pi*x)

x = np.arange(-4,5,0.01)
n = 1000

plt.title("Fourier series for x**2 on [-1,1] with the first " + str(n) + " terms") 
plt.plot(x,y(x,n)) 
plt.show()
```

```python
a_0 = 3/4
def a(n):
    if n%2 == 0:
        return 0
    else:
        return -2/(n*np.pi)**2
def b(n):
    if n%2 == 0:
        return -1/(n*np.pi)
    else:
        return 3/(n*np.pi)
def y(x : np.ndarray, n : int):
    # This is the sum of the first n terms of the fourier series.
    if n == 0:
        return np.ones(x.shape)*a_0
    else:
        return y(x,n-1) + a(n)*np.cos(n*np.pi*x) + b(n)*np.sin(n*np.pi*x)

x = np.arange(-4,5,0.1)
n = 2000

plt.title("Fourier series for 2.3.1f on [-1,1] with the first " + str(n) + " terms") 
plt.plot(x,y(x,n)) 
plt.show()
```
