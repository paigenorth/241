```python
import numpy as np 
from matplotlib import pyplot as plt 

def F(x,n,a_0,a,b):
    # This is the sum of the first n terms of the Fourier series given by a_0, a, b.
    if n == 0:
        return np.ones(x.shape)*a_0
    else:
        return F(x,n-1,a_0,a,b) + a(n)*np.cos(n*np.pi*x) + b(n)*np.sin(n*np.pi*x)
```

```python
x = np.arange(-4,5,0.01)

a_0 = 1/3
a = lambda n : 2/(n*np.pi)**2*(-1)**n
b = lambda n : 0
n = 1000

plt.title("Fourier series for x^2 on [-1,1] with the first " + str(n) + " terms") 
plt.plot(x,F(x,n,a_0,a,b)) 
plt.show()
```

```python
x = np.arange(-4,5,0.1)

a_0 = 3/4
def a(n):
    if n%2 == 0:
        return 0
    else:
        return -2/(n*np.pi)**2
def b(n):
    if n%2 == 0:
        return -1/(n*np.pi)
    else:
        return 3/(n*np.pi)
n = 100

plt.title("Fourier series for 2.3.1f on [-1,1] with the first " + str(n) + " terms") 
plt.plot(x,F(x,n,a_0,a,b)) 
plt.show()
```

```python
x = np.arange(-2,2,0.0001)

a_0 = 0
a = lambda n : 0
def b(n):
    if n%2 == 0:
        return 0
    else:
        return 4/np.pi/n
n = 1000

plt.title("Fourier series for 1 on [-1,1] with the first " + str(n) + " terms") 
plt.plot(x,F(x,n,a_0,a,b)) 
plt.show()
```

```python
x = np.arange(-2,2,0.0001)
a_0 = 0
a = lambda n : 0
def b(n):
    if n%2 == 0:
        return 2*np.pi*n*(1-np.e)/(np.pi*np.pi*n*n+1)
    else:
        return 2*np.pi*n*(1+np.e)/(np.pi*np.pi*n*n+1)
n = 2000

plt.title("Fourier series for e^x on [-1,1] with the first " + str(n) + " terms") 
plt.plot(x,F(x,n,a_0,a,b)) 
plt.show()
```
